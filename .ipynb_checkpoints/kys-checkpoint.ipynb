```python
import pandas as pd 
import numpy as np 
import tensorflow as tf 
import os 
from matplotlib import pyplot as plt 
from skimage import io 
from skimage import transform

%matplotlib inline
```

```python
nameDat = pd.read_csv('./train_labels.csv')
```

```python
datLen = len(nameDat)
splitRatio = 0.1

trainDat = nameDat[ : int(datLen * (1 - splitRatio))]
validDat = nameDat[int(datLen * (1 - splitRatio)) : int(datLen * (1 - splitRatio / 2))]
testDat = nameDat[int(datLen * (1 - splitRatio / 2)) : ]

print('elements : ' + str(datLen))
print('training : ' + str(len(trainDat)))
print('validation : ' + str(len(validDat)))
print('test : ' + str(len(testDat)))


#visualize the categories
m_train = sum(np.array(trainDat.loc[:, 'label'])) 
m_valid = sum(np.array(validDat.loc[:, 'label']))
m_test = sum(np.array(testDat.loc[:, 'label']))

b_train = - sum(np.array(trainDat.loc[:, 'label']) - 1) 
b_valid = - sum(np.array(validDat.loc[:, 'label']) - 1) 
b_test = - sum(np.array(testDat.loc[:, 'label']) - 1) 

m = [m_train, m_valid, m_test]
b = [b_train, b_valid, b_test]



index = np.arange(3)
bar_width = 0.35
opacity = 0.7

mBar = plt.barh(index + bar_width, m, bar_width, alpha = opacity, color = '#b3002d', label = 'malignant')
bBar = plt.barh(index, b, bar_width, alpha = opacity, color = '#ff668c', label = 'benign')

plt.xlabel('number of elements')
plt.yticks(index + bar_width / 2, ('Training', 'Validation', 'Test'))
plt.legend()
plt.show()
```

```python
class datGen():
    
    def mkBatch(dat, path, batch_size, ind = 0, offset = 0, n_labels = 1, aug_chance = 0.3, size = 96):
        
        imgs = []
        labels = []
        indexes = np.array(range(batch_size)) + ind * batch_size + offset
        for index in indexes:
            imgs.append(datGen.fetchImg('./%s/%s' % (path, dat.loc[index][0]), aug_chance))
            labels.append(dat.loc[index][1])
        return np.reshape(np.array(imgs) / 256, [batch_size, size * size * 3]), np.array(labels)
    
    def fetchImg(path, aug_chance):
        r = np.random.uniform()
        angle = 0
        if r < aug_chance:
            angle = np.random.uniform(90)
        return datGen.augment(io.imread('%s.tif' % (path)), angle = angle)
    
    def augment(img, angle = 0):
        if angle == 0:
            return img
        return transform.rotate(img, angle)
```

```python
elements = 10
ex_dat = datGen.mkBatch(trainDat, 'train', elements, aug_chance = 0)
```

```python
fig, ax = plt.subplots(1, elements, sharex = 'col', sharey = 'row', figsize = [10, 6], dpi = 150)
for ind, i in enumerate(ex_dat[0]):
    img = np.reshape(i, [96, 96, 3])
    ax[ind].imshow(np.mean((1-img), axis = -1), cmap = 'bone')
    ax[ind].set_title(str(ex_dat[1][ind] == 1))
```

```python
class classifier():
    
    def __init__(self, imgInput, l_rate = 0.01):
        self.x = tf.reshape(imgInput, [-1, 96, 96, 3])
        #96x96x3
        self.cnv1 = tf.layers.conv2d(self.x, 
                                     filters = 16,
                                     kernel_size = [4, 4],
                                     padding = 'same')
        self.pool1 = tf.layers.max_pooling2d(self.cnv1, 
                                             pool_size = [4, 4],
                                             strides = [2, 2],
                                             padding = 'same')
        #48x48x16
        self.cnv2 = tf.layers.conv2d(self.pool1, 
                                     filters = 32,
                                     kernel_size = [4, 4],
                                     padding = 'same')
        self.pool2= tf.layers.max_pooling2d(self.cnv2, 
                                             pool_size = [4, 4],
                                             strides = [2, 2],
                                             padding = 'same')
        #24x24x32
        self.cnv3 = tf.layers.conv2d(self.pool2, 
                                     filters = 64,
                                     kernel_size = [4, 4],
                                     padding = 'same')
        self.pool3= tf.layers.max_pooling2d(self.cnv3, 
                                             pool_size = [4, 4],
                                             strides = [2, 2],
                                             padding = 'same')
        #12x12x64
        
        self.cnv4 = tf.layers.conv2d(self.pool3, 
                                     filters = 128,
                                     kernel_size = [4, 4],
                                     padding = 'same')
        self.pool4= tf.layers.max_pooling2d(self.cnv4, 
                                             pool_size = [4, 4],
                                             strides = [2, 2],
                                             padding = 'same')
        #6x6x128
        
        self.cnv5 = tf.layers.conv2d(self.pool4, 
                                     filters = 128,
                                     kernel_size = [4, 4],
                                     padding = 'same')
        self.pool5= tf.layers.max_pooling2d(self.cnv5, 
                                             pool_size = [4, 4],
                                             strides = [2, 2],
                                             padding = 'same')
        #3x3x128
        
        self.cnv5 = tf.layers.conv2d(self.pool5, 
                                     filters = 256,
                                     kernel_size = [3, 3],
                                     padding = 'valid')
        #1x1x256
        self.flat = tf.layers.flatten(self.cnv5)
        #256
        self.XavierInit = tf.contrib.layers.xavier_initializer()
        self.fcW = tf.Variable(self.XavierInit([256, 1]))
        self.fc = tf.matmul(self.flat, self.fcW)
        #1
        self.out = tf.nn.sigmoid(self.fc)
        
        #Loss
        #Placeholder for labels 
        self.y = tf.placeholder(dtype = tf.float32, shape = [None, 1])
        self.loss = tf.nn.sigmoid_cross_entropy_with_logits(labels = self.y, logits = self.fc)
        
        #Optimize
        self.opt = tf.train.GradientDescentOptimizer(learning_rate = l_rate).minimize(self.loss)        
```

```python
#Train 
#Hyperparams 
lr = 0.0001
batch_size = 100
epochs = 10
tr_len = len(trainDat)

#create ops
tf.reset_default_graph()

flat_input = tf.placeholder(dtype = tf.float32, shape = [None, 96 * 96 * 3])

predictor = classifier(flat_input, lr)

init = tf.global_variables_initializer()

#trainable vars
tvars = tf.trainable_variables()
#saver
saver = tf.train.Saver()
```

```python
tr_losses = []
vl_losses = []
#Training loop
sess = tf.Session()
saver.restore(sess, './models/pred-0__900.ckpt')
#sess.run(init)
#saver.restore(sess, './models/pred-2__1400.ckpt')

#sess.run(init)

for epoch in range(epochs):
    for i in range(tr_len // batch_size - 1):
        temp_dat = datGen.mkBatch(trainDat, 'train', batch_size, ind = i, aug_chance = 0.1)
        temp_valid_dat = datGen.mkBatch(validDat, 'train', batch_size * 3, ind = 0, offset = tr_len + 1,aug_chance = 0.1)
            
        sess.run(predictor.opt, feed_dict = {flat_input : temp_dat[0], predictor.y : temp_dat[1].reshape([-1, 1])})
            
        tr_losses.append(sess.run(predictor.loss, feed_dict = {flat_input : temp_dat[0], predictor.y : temp_dat[1].reshape([-1, 1])}))
        vl_losses.append(sess.run(predictor.loss, feed_dict = {flat_input : temp_valid_dat[0], predictor.y : temp_valid_dat[1].reshape([-1, 1])}))
            
        if i % 100 == 0 :
            saver.save(sess, './models/pred-' + str(epoch) + '__'+ str(i) + '.ckpt')
            print('train : ' + str(np.mean(tr_losses[-1])) + ' validation : ' + str(np.mean(vl_losses[-1])) + ' ' + str(i))
```

```python
tr_pred_error = []
for i in range(55):
    print(i)
    temp_dat = datGen.mkBatch(testDat, 'train', 200, ind = i, offset = 209023, aug_chance = 0.0)
    preds = sess.run(predictor.out, feed_dict = {flat_input : temp_dat[0]}).flatten()
    preds = np.where(preds > 0.5, 1, 0)
    inaccuracy = temp_dat[1] - preds
    tr_pred_error.append(inaccuracy)

p_accu = 1 - np.sum(np.absolute(np.array(tr_pred_error).flatten())) / len(trainDat)
```
